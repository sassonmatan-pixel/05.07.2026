```python
import numpy as np
import pandas as pd 
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

## Task 1
#### Draw the lines for:
#### Add grid, legend, and axis lines x + y = 12 | x - y = 4

```python
x_value = np.arange(-20, 20 ,1)
y_value = -1 * x_value + 12
x2_value = np.arange(-20, 20 ,1)
y2_value = x_value - 4
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title('Linear function graph')
plt.plot(x_value, y_value , label = 'x + y = 12')
plt.plot(x2_value, y2_value, label = 'x - y = 4')
plt.legend()
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.show()
```

## Task 2
#### Use , , and to find the intersection point. Scatter the point and annotate it on the graphsymbolsEqsolve

```python
x_value = np.arange(-20, 20 ,1)
y_value = -1 * x_value + 12
x2_value = np.arange(-20, 20 ,1)
y2_value = x_value - 4
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title('Linear function graph')
plt.plot(x_value, y_value , label = 'x + y = 12')
plt.plot(x2_value, y2_value, label = 'x - y = 4')
plt.legend()
plt.grid(color="gray", linestyle="--", linewidth=0.5)

from sympy import symbols, Eq, solve

x = symbols("x")
y = symbols("y")

eq1 = Eq(x + y  ,12)
eq2 = Eq(x - y ,4)

solution = solve((eq1, eq2), (x, y))
print(f"The solution is: {solution}")
x_junc = solution[x]
y_junc = solution[y]
plt.scatter(x_junc, y_junc, color="blue", zorder=5)
plt.annotate(
    f"({x_junc}, {y_junc})",
    (x_junc, y_junc),
    xytext=(x_junc - 1, y_junc + 6),
    fontsize=10,
    color="blue"
)

plt.show()
```
